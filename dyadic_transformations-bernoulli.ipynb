```python
import math


def generate(start=0.03, duration=100):
    plt.axis([0, duration, 0, 1])
    y = start
    values = []
    for x in range(duration):
        values.append(y)
        y = y * 2
        if y >= 1.00:
            y = y - 1
            y = math.ceil(y*100)/100

    x = range(0, len(values))
    y = values
    return x, y
```

```python
import matplotlib.pyplot as plt
import math
start = 0.00
while start < 1.00:
    print("Generate with {} as the starting value".format(start))
    x, y = generate(start=start, duration=100)
    plt.plot(x, y, '-', linewidth=2.0)
    plt.show()
    start = start + 0.01
    start = math.ceil(start*100)/100
```

```python
import matplotlib.pyplot as plt
import math
start = 0.01
values = []
while start < 1.00:
    _, y = generate(start=start, duration=40)
    values.extend(y)
    start = start + 0.01
    start = math.ceil(start * 100)/100

x = range(len(values))
y = values
plt.plot(range(len(values)), values, '-', linewidth=2.0)
plt.show()
```
